# PROGRAMAS
2- Programa que determina si la palabra dada es palindromo.
## Definici√≥n de variables


*   P: palabra ingresada por el usuario.
*   Pi: palabra invertida.
*   Palindromo: indica si la palabra es palindromo.

## Lenguaje natural


1.   El programa pedira ingresar la palabra.
2.   Se invierte la palabra.
1.   Comparar la palabra invertida con la original.
2.   Si es palindromo mostrara que lo es; de lo contrario, dira que no es.

## Diagrama de flujo
Captura de pantalla 2025-11-10 220335.jpg

```python
p = input("Ingrese una palabra: ")
pi = p[::-1]
if p == pi:
  print("La palabra es palindromo")
else:
  print("La palabra no es palindromo.")
```
