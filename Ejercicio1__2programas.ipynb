# PROGRAMAS
## EJERCICIO 1
1- Programa que lee 3 numeros y determina cual es el mayor.

### Definicion de variables:

*   num1, num2, num3: son los numeros que ingrese el usuario.
*   m: numero mayor entre los tres.

### Descripción en lenguaje natural


1. El programa pedirá ingresar valores de num1, num2, num3.
2. Leerá y comparará valores ingresados.
1. Determinará cual es "m" (numero mayor) y mostrará el resultado.

### Diagrama de flujo


Captura de pantalla 2025-11-10 223611.jpg








```python
num1 = int(input("Ingrese el primer numero: "))
num2 = int(input("Ingrese el segundo numero: "))
num3 = int(input("Ingrese el tercer numero: "))

if num1 > num2 and num1 > num3:
  m = num1
elif num2 > num1 and num2 > num3:
  m = num2
else:
  m = num3
  print("El numero mayor es: ", m)
```
